```python
import numpy as np
import cv2
```

```python
def av_pix(img,circles,size):
    av_value = []
    for m in circles[0,:]:
        col = np.mean(img[m[1]-size:m[1]+size,m[0]-size:m[0]+size])
#         print(img[m[1]-size:m[1]+size,m[0]-size:m[0]+size])
        av_value.append(col)
    return av_value
```

```python
def get_radius(circles):
    radius = []
    for r in circles[0,:]:
        radius.append(r[2])
    return radius
```

```python
img = cv2.imread('capstone_coins.png', cv2.IMREAD_GRAYSCALE)
original_image = cv2.imread('capstone_coins.png', 1)
img = cv2.GaussianBlur(img,(5,5),0)
```

```python
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT, 0.9, 120, param1=50,param2=27,minRadius=60,maxRadius=120)
```

```python
print(circles)
```

```python
circles = np.uint16(np.around(circles))
count = 1
for i in circles[0,:]:
#     draw the outer circle
    cv2.circle(original_image,(i[0],i[1]),i[2], (0,255,0), 2)
#     draw the center of the circle
    cv2.circle(original_image,(i[0],i[1]),2,(0,0,255),3)
#     cv2.putText(original_image, str(count),(i[0],i[1]), cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,0),2)
    count += 1
```

```python
# cv2.imshow('Detected Coins', original_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
radius = get_radius(circles)
print(radius)
```

```python
bright_values = av_pix(img,circles,20)
print(bright_values)
```

```python
res = []
for i,j in zip(bright_values, radius):
    if i > 150 and j > 110:
        res.append(10)
    elif i >150 and j <=110:
        res.append(5)
    elif i < 150 and j > 110:
        res.append(2)
    elif i < 150 and j < 110:
        res.append(1)
print(res)
```

```python
count2 = 0
for i in circles[0,:]:
    cv2.putText(original_image, str(res[count2])+'p',(i[0],i[1]), cv2.FONT_HERSHEY_SIMPLEX, 2, (0,0,0), 2)
    count2 += 1
cv2.putText(original_image, 'ESTIMATED TOTAL VALUE:' + str(sum(res)) + 'p', (200,100), cv2.FONT_HERSHEY_SIMPLEX, 1.3, 255)
```

```python
cv2.imshow('Detected Coins', original_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
